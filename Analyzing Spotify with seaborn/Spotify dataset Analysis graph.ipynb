# Spotify's Dataset Analysis with seaborn

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
origin_df = pd.read_csv('data.csv')
artist_df = pd.read_csv('data_by_artist.csv')
genres_df = pd.read_csv('data_by_genres.csv')
year_df = pd.read_csv('data_by_year.csv')
```

```python
# Convert ms to minutes ans seconds
origin_df['minutes'] = origin_df['duration_ms']  / 60000
```

```python
print(origin_df['minutes'])
```

```python
# Create a bar chart visualization how long the song each year
sns.set_style('whitegrid')
sns.set_context('notebook', font_scale =1.25)
f, ax = plt.subplots(figsize = (50,10))
ax = sns.barplot(x=origin_df['year'], y =origin_df['minutes'])
```

```python
# Get the only last 20 years music duration 
# First make new DataFrame
df_2000 = origin_df[origin_df['year']>1990]


print(df_2000.head())
sns.set_style('whitegrid')
sns.set_context('notebook', font_scale =1.25)
f, ax1 = plt.subplots(figsize = (50,30))
ax1 = sns.barplot(x=(df_2000['year']), y =origin_df['minutes'])
```

```python
# comparig the minutes distribution for 20 years 

f, ax2 = plt.subplots(figsize = (50,10))
ax2 = sns.boxplot(data = df_2000, x =df_2000['year'], y=df_2000['minutes'], palette = 'Spectral')
#plt.plot(2005,2020)
plt.show()
```
